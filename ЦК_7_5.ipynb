```python
import random
input_list = [[0, 4, 8], ['a', -0], [1, 5]]

def check_input(array):
    if (len(array) == 2 and
            isinstance(array[0], int) and array[0] > 0 and
            isinstance(array[1], int) and array[1] > 0 and
            array[0] < array[1]):
        return True
    return False


def get_ladder(valid_array):
    A = valid_array[0]
    B = valid_array[1]

    N = random.randint(A, B)

    ladder1 = ['*' * i for i in range(N ,B+1)]

    ladder2 = ['*' * i for i in range(A, N + 1)]

    ladders = ladder1[::-1] +  ladder2
    return ladders


def main(input_list):
    for sub_array in input_list:
        if check_input(sub_array):
            ladder = get_ladder(sub_array)
            ans = '\n'.join(ladder)
            print(ans)
```
